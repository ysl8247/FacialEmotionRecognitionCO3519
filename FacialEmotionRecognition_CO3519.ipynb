````python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from sklearn.decomposition import PCA
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from skimage.feature import local_binary_pattern, hog
from skimage import color
from skimage.transform import swirl  # Elastic Transformations

# Function to load data from directories
def load_data(data_directory, file_extension):
    images = []
    labels = []

    if not os.path.exists(data_directory):
        print(f"Error: The directory {data_directory} does not exist.")
        return images, labels

    print(f"Loading data from {data_directory}...")
    for emotion in os.listdir(data_directory):
        emotion_path = os.path.join(data_directory, emotion)
        if not os.path.isdir(emotion_path):
            print(f"Warning: {emotion_path} is not a valid folder.")
            continue

        print(f"Loading images for {emotion}...")
        for img_name in os.listdir(emotion_path):
            if img_name.endswith(file_extension):
                img_path = os.path.join(emotion_path, img_name)
                img = cv2.imread(img_path)
                if img is None:
                    print(f"Warning: The image {img_path} could not be read.")
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
                img = cv2.resize(img, (128, 128))  # Resize to 128x128 for consistency
                images.append(img)
                labels.append(emotion)

    print(f"Loaded {len(images)} images from {data_directory}")
    return np.array(images), np.array(labels)

# Dataset paths and file formats
datasets = {
    "Cohn-Kanade": {
        "path": r'C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\Cohn-Kanade',
        "format": ".jpg"
    },
    "JAFFE": {
        "path": r'C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\JAFFE',
        "format": ".tiff"
    },
    "FacialEmotionDataset": {
        "path": r'C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\FacialEmotionDataset',
        "format": ".tiff.tiff"
    }
}

# Load all datasets
all_images = []
all_labels = []
for dataset_name, config in datasets.items():
    for data_type in ['test', 'train']:  # Loading 'test' first, then 'train' as required
        directory = os.path.join(config['path'], data_type)
        images, labels = load_data(directory, config['format'])
        all_images.extend(images)
        all_labels.extend(labels)

# Convert to numpy arrays
all_images = np.array(all_images).reshape(-1, 128, 128, 1)  # Add channel dimension
all_labels = np.array(all_labels)

# Encode labels (angry, happy, etc.)
encoder = LabelEncoder()
encoded_labels = encoder.fit_transform(all_labels)

# Normalise images to scale pixel values between 0 and 1
all_images = all_images / 255.0

# Enhanced Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=30, 
    width_shift_range=0.2, 
    height_shift_range=0.2, 
    shear_range=0.3, 
    zoom_range=0.3, 
    horizontal_flip=True, 
    fill_mode='nearest', 
    brightness_range=[0.2, 1.2]
)

datagen.fit(all_images)  # Fit the data generator on images

# Augment data (generate new images based on the original ones)
augmented_images = []
augmented_labels = []
for _ in range(5):  # Number of augmented images per original image
    augmented_batch = datagen.flow(all_images, encoded_labels, batch_size=len(all_images), shuffle=False)
    augmented_images_batch, augmented_labels_batch = next(augmented_batch)
    augmented_images.extend(augmented_images_batch)
    augmented_labels.extend(augmented_labels_batch)

augmented_images = np.array(augmented_images)
augmented_labels = np.array(augmented_labels)

# Elastic Transformations for further augmentation
def elastic_transform(image, alpha, sigma):
    random_state = np.random.RandomState(None)
    shape = image.shape
    dx = random_state.uniform(-1, 1, shape) * alpha
    dy = random_state.uniform(-1, 1, shape) * alpha
    dz = np.zeros_like(dx)
    distorsion = np.dstack([dx, dy, dz])

    # Ensure kernel size is odd
    kernel_size = int(2 * np.floor(sigma) + 1)
    distorsion = cv2.GaussianBlur(distorsion, (kernel_size, kernel_size), 0)
    
    # Convert the distortion maps to the correct type (CV_32FC2)
    map1, map2 = cv2.convertMaps(distorsion[..., 0].astype(np.float32), 
                                  distorsion[..., 1].astype(np.float32), 
                                  cv2.CV_32F)

    # Apply the remapping with elastic transformation
    distorted_image = cv2.remap(image, map1, map2, interpolation=cv2.INTER_LINEAR)
    return distorted_image

# Apply elastic transformations to augmented images
augmented_images_elastic = []
for img in augmented_images:
    img_elastic = elastic_transform(img.squeeze(), alpha=34, sigma=4)  # Alpha and sigma for elastic transformations
    augmented_images_elastic.append(np.expand_dims(img_elastic, axis=-1))  # Add channel dimension back

augmented_images = np.array(augmented_images_elastic)
augmented_labels = np.array(augmented_labels)

# Feature extraction - HOG (Histogram of Oriented Gradients)
def extract_hog_features(images):
    hog_features = []
    for img in images:
        img = img.squeeze()  # Ensure image is 2D
        fd = hog(img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=False)
        hog_features.append(fd)
    return np.array(hog_features)

# Feature extraction - LBP (Local Binary Patterns)
def extract_lbp_features(images, radius=1, n_points=8):
    lbp_features = []
    for img in images:
        img = img.squeeze()  # Ensure image is 2D
        img = img.astype(np.uint8)  # Convert image to integer type (uint8)
        lbp = local_binary_pattern(img, P=n_points, R=radius, method="uniform")
        lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
        lbp_hist = lbp_hist.astype("float")
        lbp_hist /= lbp_hist.sum()  # Normalise the histogram
        lbp_features.append(lbp_hist)
    return np.array(lbp_features)

# Use both HOG and LBP for feature extraction
hog_features = extract_hog_features(augmented_images)
lbp_features = extract_lbp_features(augmented_images)

# Combine both HOG and LBP features for training
features = np.hstack([hog_features, lbp_features])

# Apply PCA (Principal Component Analysis) for dimensionality reduction (to retain 95% of variance)
pca = PCA(n_components=0.95)
features_pca = pca.fit_transform(features)

# Define classifiers (SVM, Decision Tree, KNN, Naive Bayes)
classifiers = {
    'SVM': SVC(kernel='linear', C=1, class_weight='balanced'),  # Class weight to handle imbalance
    'Decision Tree': DecisionTreeClassifier(max_depth=10),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}

# Hyperparameter tuning with GridSearchCV
param_grid = {
    'SVM': {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf'], 'gamma': ['scale', 'auto'], 'class_weight': ['balanced', None]},
    'Decision Tree': {'max_depth': [5, 10, 15], 'min_samples_split': [2, 5, 10]},
    'KNN': {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']},
    'Naive Bayes': {'var_smoothing': [1e-9, 1e-8, 1e-7]}
}

# Apply GridSearchCV for hyperparameter tuning
grid_search = {}
for name, clf in classifiers.items():
    print(f"Hyperparameter tuning for {name}...")
    grid_search[name] = GridSearchCV(clf, param_grid[name], cv=5)
    grid_search[name].fit(features_pca, augmented_labels)

# K-Fold Cross-validation (5 folds) integrated into the training process
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Store accuracy for plotting
accuracies = {'train': [], 'test': []}

for name, clf in classifiers.items():
    print(f"Training {name}...")
    for train_idx, val_idx in kf.split(features_pca):
        X_train, X_val = features_pca[train_idx], features_pca[val_idx]
        y_train, y_val = augmented_labels[train_idx], augmented_labels[val_idx]
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_val)

        print(f"{name} Classification Report:")
        print(classification_report(y_val, y_pred, target_names=encoder.classes_, zero_division=1))  # Zero division set to 1 to avoid warnings

        # Store accuracy for plotting
        accuracies['train'].append(clf.score(X_train, y_train))
        accuracies['test'].append(clf.score(X_val, y_val))

        # Confusion Matrix
        cm = confusion_matrix(y_val, y_pred)
        sns.heatmap(cm, annot=True, cmap='Blues', xticklabels=encoder.classes_, yticklabels=encoder.classes_)
        plt.title(f'{name} Confusion Matrix')
        plt.show()

# Plot accuracy graphs for train and test sets
plt.figure(figsize=(10, 6))
plt.plot(accuracies['train'], label='Training Accuracy')
plt.plot(accuracies['test'], label='Testing Accuracy')
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Training and Testing Accuracy over Cross-Validation Folds')
plt.legend(loc='best')
plt.show()

# Preprocess and flatten the image for PCA transformation
def preprocess_for_pca(img, pca):
    img_resized = cv2.resize(img, (128, 128))  # Resize image to match input size
    img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    img_flattened = img_gray.reshape(1, -1)  # Flatten the image
    if img_flattened.shape[1] != pca.components_.shape[1]:  # Check for dimensionality mismatch
        img_flattened = img_flattened[:, :pca.components_.shape[1]]  # Resize features to match PCA
    img_pca = pca.transform(img_flattened)  # Apply PCA transformation
    return img_pca

# Test images
test_images = [
    r"C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\Cohn-Kanade\test\happy\happy_image1.jpg",
    r"C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\JAFFE\test\sad\sad_image1.tiff",
    r"C:\Users\HP\OneDrive\Desktop\FacialEmotionRecognition_CO3519\FacialEmotionDataset\test\angry\angry_image1.tiff.tiff"
]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for idx, path in enumerate(test_images):
    img = cv2.imread(path)
    if img is None:
        print(f"Error loading {path}")
        continue
    
    # Preprocess image for PCA transformation
    img_pca = preprocess_for_pca(img, pca)

    # Predict with all classifiers
    predictions = {name: clf.predict(img_pca)[0] for name, clf in classifiers.items()}

    # Display predictions
    axes[idx].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert to RGB for display
    axes[idx].set_title("\n".join([f"{k}: {encoder.inverse_transform([v])[0]}" for k, v in predictions.items()]))
    axes[idx].axis('off')

plt.tight_layout()
plt.show()
````
